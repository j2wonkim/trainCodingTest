```python
# 2023/02/25
# 백준, 1325번.효율적인 해킹
# https://www.acmicpc.net/problem/1325

def effective_hacking():
    N, M = map(int, input().split())
    graph = {i:[] for i in range(1, N + 1)}
    to_find_graph = dict()
    print(graph)
    
    for i in range(M):
        value, node = map(int, input().split())
        graph[node].append(value)
    
    print(graph)
    
    for key, value in graph.items():
        if len(value) > 0:
            to_find_graph[key] = value
    
    dfs(to_find_graph)

def dfs(graph):    
    key_list = list(graph.keys())
    key_list.sort()
    print(key_list)
    
    result_graph = {i:[] for i in key_list}
    print("result_graph:", result_graph)
    result_count = list()
        
    for key in key_list:
        need_visit, visited = list(), list()
        count = 0
        
        need_visit.append(key)
        print("key:", key)

        while need_visit:
            count += 1
            node = need_visit.pop()
            print("node:", node)

            if node not in visited:
                visited.append(node)
                print("visited:", visited)
                if node in key_list:
                    need_visit.extend(graph[node])
        
        result_graph[key] = visited
        print(result_graph)
        result_count.append([count, key])
        print("result_count:", result_count)
    
    result_count.sort(reverse=True)
    print("정렬 후 result_count:", result_count)
    
    result_list = list()
    for i in range(len(result_count)):
        if result_count[0][0] == result_count[i][0]:
            result_list.append(result_count[i][1])
    
    result_list.sort()    
    
    print(result_list)
```

```python
effective_hacking()
```
